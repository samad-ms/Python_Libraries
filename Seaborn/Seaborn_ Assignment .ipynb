Scatter Plot:
Use: Visualize the relationship between two numeric variables.
Example: Analyzing the correlation between a person's age and their income.

Bar Plot:
Use: Display and compare categorical data.
Example: Showing the sales performance of different products in a store.

Histogram:
Use: Explore the distribution of a single numeric variable.
Example: Examining the frequency distribution of exam scores in a class.

Heatmap:
Use: Visualize a matrix of data with color-coded values.
Example: Representing the correlation matrix of financial stock returns.

Box Plot (Box and Whisker Plot):
Use: Summarize the distribution and identify outliers in numeric data.
Example: Analyzing the distribution of housing prices in a neighborhood.

### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =  "timepoint" and y = "signal" for different events and regions.  
Note:  timepoint, signal, event, and region are columns in the fmri dataset. 

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fmri=sns.load_dataset('fmri')
fmri.head()
```

```python
sns.relplot(x='timepoint',y='signal',data=fmri,kind='line',hue='event',style='region')
# Add labels and a legend
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("FMRI Signal Over Time by Event and Region")
plt.figure(figsize=(10,6))
plt.show()
```

### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =  'pclass', y = 'age' and y = 'fare'. 


```python
titanic=sns.load_dataset('titanic')
titanic
```

```python
# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")
# Create subplots with two box plots
plt.figure(figsize=(12, 6))

# Box plot for 'pclass' vs. 'age'
plt.subplot(1, 2, 1)
sns.boxplot(x="pclass", y="age", data=titanic_data)
plt.title("Box Plot of Age by Pclass")

# Box plot for 'pclass' vs. 'fare'
plt.subplot(1, 2, 2)
sns.boxplot(x="pclass", y="fare", data=titanic_data)
plt.title("Box Plot of Fare by Pclass")

# Adjust layout
plt.tight_layout()
# Show the plots
plt.show()
```

### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue  parameter for the 'cut' column of the diamonds dataset.  


```python
diamonds=sns.load_dataset('diamonds')
diamonds.head()
```

```python
sns.histplot(data=diamonds, x="price", hue="cut", bins=30, kde=True)
# Add labels and a title
plt.xlabel("Price")
plt.ylabel("Count")
plt.title("Histogram of Diamond Prices by Cut")
plt.show()
```

### Que 6: Use the "flights" dataset from seaborn to plot a heatmap. 

```python
flights=sns.load_dataset('flights')
flights.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Pivot the dataset to create a matrix suitable for a heatmap
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')
# Create a heatmap
plt.figure(figsize=(10, 8))  # Optional: Adjust the figure size
sns.heatmap(flights_pivot, cmap='coolwarm', annot=True, fmt='d')

# Add labels and a title
plt.xlabel("Year")
plt.ylabel("Month")
plt.title("Passenger Count Heatmap (Flights Dataset)")
plt.show()
```
